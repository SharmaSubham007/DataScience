```python
x = pd.Series([24.23,25.53,25.41,24.14,29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.71,35.00])
company_name = pd.Series(['Allied Signal','Bankers Trust','General Mills','ITT Industries','J.P.Morgan & Co.','Lehman Brothers',
      'Marriott','MCI','Merrill Lynch','Microsoft','Morgan Stanley','Sun Microsystems','Travelers','US Airways',
      'Warner-Lambert'])
```

```python
plt.figure(figsize=(6,6))
plt.bar(company_name,x) # Plotting the data in bar graph
plt.xticks(rotation=90) # Rotating the names on the X-axis
plt.show()
```

```python
plt.boxplot(x)
```

```python
x.describe()
```

```python
q1 = 25.47    # we can also use np.percentile(x,25)
q3 = 33.975   # we can also use np.percentile(x,75)
iqr = q3-q1   # Interquartile range
lower = q1 - 1.5 * iqr # Lower boundary
upper = q3 + 1.5 * iqr # Upper boundary
outliers  = [outlier for outlier in x if outlier<=lower or outlier>=upper]
print("The outliers in the given data are:",outliers)
```
